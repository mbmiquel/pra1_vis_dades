```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("wine_dataset_merged.csv")
```

```python
# Totes les paraules en una variable
text = " ".join(description for description in df.description)

# Eliminem algunes paraules comuns
common_words = ['the', 'and', 'is', 'a', 'an', 'but', 'that', 'are', 'this', 'in', 'also', 'now', 'to', 'be', 'it', 'has']
for word in common_words:
    text = text.replace(" " + word + " ", " ")

# Generem el nuvol de paraules
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = set(common_words),
                min_font_size = 10).generate(text)

# Plot                       
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```

```python
# Guardar com a fitxer
wordcloud.to_file("word_cloud.png")
```
