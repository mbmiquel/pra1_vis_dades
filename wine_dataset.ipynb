```python
import pandas as pd
```

*1. Carreguem les dades*

```python
df_wines = pd.read_csv("winemag-data-130k-v2.csv")
df_wines.drop("Unnamed: 0", axis = 1, inplace = True)
df_wines
```

```python
df_countries = pd.read_csv("country-codes_csv.csv")
df_countries
```

```python
# Filtrem el dataset de countries per aquelles columnes interesants per a l'anàlisis
df_countries = df_countries[["official_name_en", "Region Name", "Sub-region Name", "Developed / Developing Countries"]]
# df_countries = df_countries.loc[~df_countries["official_name_en"].isnull()]
df_countries
```

*2. Unim els dos datasets*

```python
# S'ha detectat que alguns noms de països no estan escrits de la mateixa manera, per tant, caldrà homogeneïtzar els valors
countries_b = set(df_countries["official_name_en"].unique())
countries_a = set(df_wines["country"].unique())
countries_a - countries_b
```

```python
replace_dict = {"Czechia" : "Czech Republic",
                "United Kingdom of Great Britain and Northern Ireland" : "England",
                "The former Yugoslav Republic of Macedonia" : "Macedonia",
                "Republic of Moldova" : "Moldova",
                "United States of America" : "US"}
df_countries.loc[df_countries['official_name_en'].isin(replace_dict.keys()), 'official_name_en'] = df_countries['official_name_en'].map(replace_dict)
df_countries['official_name_en']
```

```python
df_wines_merged = pd.merge(df_wines, df_countries, how="left", left_on = ["country"], right_on = ["official_name_en"])
df_wines_merged
```
